```python
from scipy.signal import find_peaks
import matplotlib.pyplot as plt
```

```python
import numpy as np
spike=np.array([])
spike=[1,3,5,7,9,11,5,7,3,1,7]
plt.plot(spike)
```

```python
peaks,properties=find_peaks(spike,height=0.1)
print('index of peak',peaks)
print('height of each peaks',properties["peak_heights"])
plt.plot(spike)
plt.plot(peaks, properties["peak_heights"], "x")
```

```python
# Hight Threshold
peaks,properties=find_peaks(spike,height=8)
print('index of peak',peaks)
print('height of each peaks',properties["peak_heights"])
plt.plot(spike)
plt.plot(peaks, properties["peak_heights"], "x")
```
